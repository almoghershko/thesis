# Loading Data

```python
# load the RF's distance matrix
load_RF_name = 'simple___2021_11_27___22_09_00___standard_RF_max_depth_10'
s3_load_dir_path = os.path.join(s3_saves_dir_path, 'RF', load_RF_name)
sim_mat = from_s3_npy(s3_client, bucket_name, os.path.join(s3_load_dir_path, 'sim_mat.npy'))
dis_mat = 1 - sim_mat
```

```python
# taking the training part out of it
from sklearn.model_selection import train_test_split
I_train_NN, I_test_NN = train_test_split(np.arange(dis_mat.shape[0]), train_size=9000, random_state=seed)
dis_mat = dis_mat[I_train_NN,:][:,I_train_NN]
D = dis_mat[np.triu_indices(9000)] # only the upper triangular part, as the matrix is symmetric
```

```python
# loading the NNs distance matrix (upper triangular part only)
D_NN = from_s3_npy(s3_client, bucket_name, os.path.join(s3_saves_dir_path, 'NN', 'simple___2021_12_06___21_36_21___Kernels_31_Filters_64_32_16_8_4_Hiddens_512_128_tanh','Z.npy'))
D_NN_denoise = from_s3_npy(s3_client, bucket_name, os.path.join(s3_saves_dir_path, 'NN', 'simple___2021_12_13___11_16_51___Kernels_31_Filters_64_32_16_8_4_Hiddens_512_128_tanh_SNR_6dB_40dB','Z.npy'))
#D_NN = from_s3_npy(s3_client, bucket_name, os.path.join(s3_saves_dir_path, 'NN', 'LongTrain___2022_01_18___12_39_15___Kernels_31_Filters_64_32_16_8_4_Hiddens_512_128_tanh','Z.npy'))
#D_NN_denoise = from_s3_npy(s3_client, bucket_name, os.path.join(s3_saves_dir_path, 'NN', 'LongTrainDenoise___2022_01_18___12_39_30___Kernels_31_Filters_64_32_16_8_4_Hiddens_512_128_tanh','Z.npy'))
```

# Error Histogram

```python
# calculating the error
D_NN_err = abs(D_NN-D)
D_NN_denoise_err = abs(D_NN_denoise-D)
```

```python
# calculating histograms
N = 1000
edges = np.linspace(0,1,N+1)
centers = (edges[:-1]+edges[1:])/2
err_NN_counts,_ = np.histogram(D_NN_err, bins=edges)
err_NN_denoise_counts,_ = np.histogram(D_NN_denoise_err, bins=edges)
err_NN_counts = np.divide(err_NN_counts, np.sum(err_NN_counts)) # normalizing to get PDF (the "normed=True" option in np.histogram is broken)
err_NN_denoise_counts = np.divide(err_NN_denoise_counts, np.sum(err_NN_denoise_counts)) # normalizing to get PDF (the "normed=True" option in np.histogram is broken)
```

```python
# plotting
hist_fig, hist_ax = plt.subplots(figsize=(15,8))
hist_ax.set_title('Error PDF (L1 between the distances)')
hist_ax.set_xlabel('err')
hist_ax.set_ylabel('Pr')
hist_ax.grid()
hist_ax.plot(centers,err_NN_counts,label='NN')
hist_ax.plot(centers,err_NN_denoise_counts,label='NN denoise')
hist_ax.legend()
```

```python
hist_fig, hist_ax = plt.subplots(figsize=(15,8))
hist_ax.set_title('Error PDF (L1 between the distances) - Zoom In')
hist_ax.set_xlabel('err')
hist_ax.set_ylabel('freq')
hist_ax.grid()
hist_ax.plot(centers[:200],err_NN_counts[:200],label='NN')
hist_ax.plot(centers[:200],err_NN_denoise_counts[:200],label='NN denoise')
hist_ax.legend()
```

```python
hist_fig, hist_ax = plt.subplots(figsize=(15,8))
hist_ax.set_title('Error PDF (L1 between the distances)')
hist_ax.set_xlabel('err')
hist_ax.set_ylabel('freq')
hist_ax.grid()
hist_ax.plot(centers,err_NN_counts,label='NN')
hist_ax.plot(centers,err_NN_denoise_counts,label='NN denoise')
hist_ax.legend()
hist_ax.set_yscale('log', base=2)
```

## Normalizing with Distances PDF to get the conditional PDF: Pr(err|dist)=Pr(err^dist)/Pr(dist)

```python
H,_,_ = np.histogram2d(D_NN_err, D, bins=(edges,edges)) # calculating Pr(err^dist)
H = H.T
D_counts,_ = np.histogram(D, bins=edges) # calculating Pr(dist)
Pr_err_given_dist = np.divide(H,np.repeat(D_counts.reshape(-1,1),H.shape[1],1)) # calculating Pr(err|dist)
Pr_err_given_dist = np.nan_to_num(Pr_err_given_dist)
```

```python
Pr_err_given_dist = np.divide(Pr_err_given_dist, np.repeat(np.sum(Pr_err_given_dist, axis=1).reshape(-1,1), Pr_err_given_dist.shape[1],1))
Pr_err_given_dist = np.nan_to_num(Pr_err_given_dist)
```

```python
# plotting
hist_fig, hist_ax = plt.subplots(figsize=(15,8))
hist_ax.set_title('Pr(Error|Distance) - NN (no denoise)')
hist_ax.set_xlabel('error')
hist_ax.set_ylabel('distance')
pos = hist_ax.imshow(Pr_err_given_dist, interpolation='nearest', origin='lower', extent=[edges[0], edges[-1], edges[0], edges[-1]], vmin=Pr_err_given_dist.min(), vmax=Pr_err_given_dist.max())
hist_fig.colorbar(pos, ax=hist_ax)
```

```python
# plotting
hist_fig, hist_ax = plt.subplots(figsize=(15,8))
hist_ax.set_title('EPr(Error|Distance) - NN (no denoise) - log scale')
hist_ax.set_xlabel('error')
hist_ax.set_ylabel('distance')
pos = hist_ax.imshow(Pr_err_given_dist, interpolation='nearest', origin='lower', extent=[edges[0], edges[-1], edges[0], edges[-1]], norm=LogNorm())
hist_fig.colorbar(pos, ax=hist_ax)
```

# TopK

```python
# calculating weird scores for RF
weird_scores = np.mean(dis_mat, axis=1)

# calculating full distance matrix for NN
D_NN_mat = np.zeros_like(dis_mat)
D_NN_mat[np.triu_indices(dis_mat.shape[0])] = D_NN
D_NN_mat = D_NN_mat.T
D_NN_mat[np.triu_indices(dis_mat.shape[0])] = D_NN

# calculating weirdness scores for NN
NN_weird_scores = np.mean(D_NN_mat, axis=1)

# calculating full distance matrix for NN
D_NN_denoise_mat = np.zeros_like(dis_mat)
D_NN_denoise_mat[np.triu_indices(dis_mat.shape[0])] = D_NN_denoise
D_NN_denoise_mat = D_NN_denoise_mat.T
D_NN_denoise_mat[np.triu_indices(dis_mat.shape[0])] = D_NN_denoise

# calculating weirdness scores for NN
NN_denoise_weird_scores = np.mean(D_NN_denoise_mat, axis=1)
```

```python
# calculating ranks
RF_rank = np.argsort(weird_scores)[::-1]
NN_rank = np.argsort(NN_weird_scores)[::-1]
NN_denoise_rank = np.argsort(NN_denoise_weird_scores)[::-1]

# calculating match
NN_match = [len(set(RF_rank[:(k+1)]).intersection(set(NN_rank[:(k+1)])))/(k+1) for k in range(len(RF_rank))]
NN_denoise_match = [len(set(RF_rank[:(k+1)]).intersection(set(NN_denoise_rank[:(k+1)])))/(k+1) for k in range(len(RF_rank))]
```

```python
# plotting match
rank_fig, rank_ax = plt.subplots(figsize=(15,8))
rank_ax.plot(np.arange(1,len(RF_rank)+1), NN_match, label='NN')
rank_ax.plot(np.arange(1,len(RF_rank)+1), NN_denoise_match, label='NN denoise')
rank_ax.legend()
rank_ax.set_title('TopK Match Ratio')
rank_ax.set_xlabel('K')
rank_ax.set_ylabel('Match Ratio')
```

```python
# plotting match
rank_fig, rank_ax = plt.subplots(figsize=(15,8))
N = 200
rank_ax.plot(np.arange(1,N+1), NN_match[:N], label='NN')
rank_ax.plot(np.arange(1,N+1), NN_denoise_match[:N], label='NN denoise')
rank_ax.legend()
rank_ax.set_title('TopK Match Ratio')
rank_ax.set_xlabel('K')
rank_ax.set_ylabel('Match Ratio')
```
